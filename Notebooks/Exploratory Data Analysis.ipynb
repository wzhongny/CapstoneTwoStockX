```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df = pd.read_csv('sneaker_data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
#drop unnamed column
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
#Create bar graph to show number of shoes sold for each size
shoe_sizes = df['Shoe Size'].value_counts().sort_index().plot(kind='bar')
plt.title('Sales by Shoe Size')
plt.xlabel('Size')
plt.ylabel('Sales')
plt.show()
```

```python
#Let's find out if the return on investment differs significantly between the two brands
#Compute the difference in means first
yz_mean = np.mean(df[df['Brand']=='Yeezy']['ROI'])
ow_mean = np.mean(df[df['Brand']=='Off-White']['ROI'])
print('Mean return on investment for Yeezy: ', yz_mean)
print('Mean return on investment for Off-White: ', ow_mean)
obs_diff = abs(yz_mean - ow_mean)
print('Difference of the two means: ', obs_diff)
```

```python
#There is a difference of about 200% which looks significant
#Null hypothesis: No difference in means
#Alternaltive hypothesis: Means are significantly different
#Create a boxplot to visualize the difference
df.boxplot(by='Brand', column='ROI', grid=False, rot=45, fontsize=15)
```

```python
#We can see there is a big difference there
#Now let's see the distribution on each one
plt.hist(df[df['Brand']=='Yeezy']['ROI'])
plt.title('Yeezy ROI distribution')
plt.xlabel('ROI')
plt.ylabel('Sales')
plt.show()
```

```python
plt.hist(df[df['Brand']=='Off-White']['ROI'])
plt.title('Off-White ROI distribution')
plt.xlabel('ROI')
plt.ylabel('Sales')
plt.show()
```

```python
#Distributions are not normal, we are going to perform a permutation test
np.random.seed(350)
difference = []

for i in range(10000):
    permutation = np.random.permutation(df['ROI'])
    difference.append(abs(np.mean(permutation[df['Brand']=='Yeezy']) - np.mean(permutation[df['Brand']=='Off-White'])))
```

```python
#Plot histogram of the difference list
plt.hist(difference)
plt.title('Distribution of difference in means of ROI')
plt.xlabel('Difference in means')
plt.ylabel('Number of outcomes')
plt.show()
```

```python
#Find p-value
p = np.sum(np.array(difference) >= obs_diff) / len(difference)
p
```

```python
#P-value is 0 which means we reject our null hypothesis
#We conclude that the two brands have significantly different returns on investment, Off-White being significantly higher than Yeezy
#Here, I'm going to make 'Brand' a binary variable with Yeezy represented by 0 and Off-White represented by 1 so that we can incorporate it into our regression model later
df[['Brand']] = df[['Brand']].replace(['Yeezy'], 0)
df[['Brand']] = df[['Brand']].replace(['Off-White'], 1)
```

```python
#Remove 'days' from Days_after_release column and make it an integer
df[['Days_after_release']] = df['Days_after_release'].str.replace(' days', '').astype(int)
df.info()
```

```python
#Remove non-numerical columns, as well as 'Sale Price' column (using ROI column)
df = df.drop(columns=['Order Date', 'Sneaker Name', 'Sale Price', 'Release Date', 'Buyer Region'])
df.head()
```

```python
#Create pairplot of the data
sns.pairplot(df)
plt.show()
```

```python
#Create a heatmap to see the correlation between each feature
sns.heatmap(df.corr(), annot=True)
plt.show()
```

Return on investment(ROI) is our dependent variable. Looking at the heatmap, it looks as though brand and retail price have the most effect on ROI, while shoe size and days after release have the least effect. We can use these variables to create a regression model for ROI.
