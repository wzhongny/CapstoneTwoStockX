```python
#Import pandas, matplotlib.pyplot, and datetime
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
#Read in the csv file
sneaker_data = pd.read_csv('../raw_data/StockX-Data-Contest-2019-3.csv')
```

```python
#Look at summary of dataset
sneaker_data.info()
```

```python
#There are 99956 entries and no missing data in the dataset
#Date columns should be changed to date type and prices columns should set to int
#Call head to print first rows of dataset
sneaker_data.head()
```

```python
#Remove leading and trailing spaces in string columns
sneaker_data['Brand'] = sneaker_data['Brand'].str.strip()
sneaker_data['Sneaker Name'] = sneaker_data['Sneaker Name'].str.strip()
sneaker_data['Buyer Region'] = sneaker_data['Buyer Region'].str.strip()
```

```python
#Change the data types for the date columns and the price columns
sneaker_data[['Order Date', 'Release Date']] = sneaker_data[['Order Date', 'Release Date']].apply(pd.to_datetime)
#Removing $ and , before changing price to type int
sneaker_data['Sale Price'] = sneaker_data['Sale Price'].str.replace('$', '', regex=False)
sneaker_data['Sale Price'] = sneaker_data['Sale Price'].str.replace(',', '', regex=False)
sneaker_data['Retail Price'] = sneaker_data['Retail Price'].str.replace('$', '', regex=False)
sneaker_data[['Sale Price', 'Retail Price']] = sneaker_data[['Sale Price', 'Retail Price']].apply(pd.to_numeric)
sneaker_data.info()
```

```python
#Look to see if there are order dates that come before release date
sneaker_data.loc[sneaker_data['Order Date'] < sneaker_data['Release Date']]
```

```python
#5601 pairs of sneakers were sold before the release date
#Removing these entries because they are irrelevant to our project
sneaker_data = sneaker_data.loc[sneaker_data['Order Date'] > sneaker_data['Release Date']]
```

```python
sneaker_data.head()
```

```python
#Check average, ranges, and potential outliers
sneaker_data.describe()
```

```python
#Big range for sale price
#Graph out to see distribution
fig, ax = plt.subplots(figsize = (12,8))
plt.hist(sneaker_data['Sale Price'])
ax.set_title('Sale Price')
ax.set_xlabel('Price')
ax.set_ylabel('Count')
```

```python
#Most sale prices are in 0-500 range
sneaker_data.loc[sneaker_data['Sale Price'] > 4000]
```

```python
sneaker_data.loc[(sneaker_data['Sale Price'] > 3000) & (sneaker_data['Sale Price'] < 4000)]
```

```python
#All the sneakers that resaled above $3000 are all from the Off-White brand
#Separate the two brands and see sale price distribution
off_white = sneaker_data.loc[sneaker_data['Brand'] == 'Off-White']
off_white.describe()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.hist(off_white['Sale Price'])
ax.set_title('Off-White')
ax.set_xlabel('Price')
ax.set_ylabel('Count')
```

```python
#A useful statistic to have for this would be how many pairs were released for retail to determine how limited
#each shoe is
#Checking with StockX, these outliers are correct and not errors
#Create new column, days after release date
sneaker_data['Days_after_release'] = sneaker_data['Order Date'] - sneaker_data['Release Date']
sneaker_data.head()
```

```python
#Create column, return on investment
#There are taxes when buying from retail as well as fees for selling on StockX, so won't be exact
sneaker_data['ROI'] = (sneaker_data['Sale Price'] - sneaker_data['Retail Price'])/sneaker_data['Retail Price'] * 100
sneaker_data.head()
```
